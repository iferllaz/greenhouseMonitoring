## **Analysis**

The Measures of Central Tendency play a significant role in analyzing the data obtained from an agent IoT simulator that records the behavior of a greenhouse's temperature, humidity, luminosity, soil moisture, water pH, and soil pH. These measures provide a clear understanding of the distribution of the data and the central values around which the data is clustered.

The greenhouse cultivates a variety of vegetables and with the help of the IoT agent the team was able to get a dataset containing relevant information about the grenhouse's environment for the purpose of analyzing and optimizing the growing conditions of these crops.

The dataset contains information about the temperature, humidity, light, soil moisture, water acidity, and soil acidity inside the greenhouse. Our team of data scientists will analyze this data to find ways to improve the conditions for growing vegetables.

As a team, we adhere to the recommended values for the vegetable greenhouse, including maintaining appropriate temperature, humidity, luminosity, soil moisture, water pH, and soil pH. However, we also understand that every crop has unique needs, and we continuously monitor and adjust the conditions of our greenhouse to ensure that each plant receives the best possible growing conditions.

The recommended values for a vegetable greenhouse are:

- **Temperature**: between 18C to 27C (64.4F to 78.8F)
- **Humidity**: between 50% to 70%
- **Luminosity**: generally, vegetables require between 12 to 16 hours of light per day for optimal growth.
- **Soil moisture**: between 50% to 80%
- **Water pH**: between 5.5 to 6.5
- **Soil pH**: between 6.0 to 7.0

However, the specific values may vary depending on the type of vegetable being grown, the growth stage, and the specific conditions of the greenhouse.


Having understood the importance of the environment of a greenhouses, we can now proceed to explain the process of analyzing the dataset obtained from the IoT simulator using Python.
